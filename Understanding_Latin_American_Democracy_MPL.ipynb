# **1. Problem Statement**



Democracy, is a contested word that has no true definition that can serve as a [panacea](https://books.google.com/books?hl=en&lr=&id=uiFH5dh12p0C&oi=fnd&pg=PR11&dq=democracy+and+development+adam+przeworski&ots=xmja4TCxkx&sig=TJb92nki_wdZfoHMsesvRPGJxjg#v=onepage&q=democracy%20and%20development%20adam%20przeworski&f=false). However, one generally accepted definition is that democracy protects the rights of minorities and institutionalizes [uncertainty](https://books.google.com/books?hl=en&lr=&id=6wJvFVyhKJwC&oi=fnd&pg=PP1&dq=juan+linz+and+schmitter+transitions+from+authoritarianism&ots=2xIU-WQ0Pm&sig=tycVeJPCIecSWW2KuRgWfXT0FMQ#v=onepage&q=juan%20linz%20and%20schmitter%20transitions%20from%20authoritarianism&f=false). It does so in the following way: although the majority may be in power today, through free and fair elections, there is always the chance that the minority could come back to take power. This is what we mean by protecting the rights of minorities -- the majority is not able to stack the deck in their hands to ensure that the minority has no chance at winning or getting their policies passed through. This form of democracy, what we call liberal democracy, is under threat from authoritarianism and democratic backsliding like never before. Even the United States – the world's longest standing democracy -- is now categorized as a [democracy in decline.](https://www.idea.int/blog/democratic-backsliding-different-causes-divergent-trajectories)


This democratic backsliding is having grave implications for world freedom and prosperity as free and fair elections – a pivotal component of democracy – [are being undermined](https://scholar.harvard.edu/levitsky/files/SL_elections.pdf). This is causing less and less people to ultimately have a voice and or say in matters that impact them and their daily lives.

This democratic backsliding and shift towards authoritarianism is caused by multiple mechanisms. For one, backsliding occurs on behalf of state actors that [deliberately undermine](https://www.idea.int/blog/democratic-backsliding-different-causes-divergent-trajectories) democratic political institutions in a given country. Actions that damage horizontal constraints in a given country are also a causal mechanism for the decline in democracy such as attacks on the legislature and the courts along with proposed laws and measures that challenge election and media integrity – along with the degradation of civil liberties. Another huge cause of backsliding is the concept of [‘majoritarianism’](https://igcc.ucsd.edu/news-events/news/why-is-democratic-backsliding-on-the-rise.html) whereby the majority who win in an election think they have an extreme mandate to rule without cause for concern for the losers in an election. Polarization between citizens only [exacerbates](https://igcc.ucsd.edu/news-events/news/why-is-democratic-backsliding-on-the-rise.html) the backsliding trend. 


As a current example of the dangers of democracy in decline - and if we need any more look that democracy is [under attack and in decline](https://freedomhouse.org/country/venezuela/freedom-world/2021), simply look at the case of Venezuela (a country pertinent to our analyses) over the years. Venezuela was once a model democracy but has since experienced a deep democratic decline to the point where the country is now a full blown authoritarian state. This has implications not only for the country's institutions such as its court system and integrity of the rule of law, but also weakens civil society and arguably makes the population worse off. 

However, this trend in democratic backsliding may not be the same in all places world wide. Perhaps some regions of the world are sliding into democratic regress whereas others may be strengthening their democratic foundations. In order to analyze this, we pull from the [Varieties of Democracy index](http://v-dem.net/vdemds.html) using the liberal democracy indicator that measures the extent of liberal democracy of a given country on a normalized scale ranging from 0-1. By liberal democracy – the measure refers to the extent surrounding the importance of protecting minority rights against tyranny of the majority. In liberal democracies, this is achieved via constitutional means that protect civil liberties, the rule of law including a strong and independent judiciary, and effective checks and balances that limit the exercise of executive power.

Overall, this trend worldwide is concerning because we also know that people are generally [more satisfied and happy](https://sites.psu.edu/tetirclblog/2015/10/12/the-correlation-between-democracy-and-happiness/) living in democracies. Thus, the policy implications of democratic rule cannot be discounted as people’s general well-being is much more robust under democratic rule relative to authoritarian rule.

Together with the changes in democracy, the other indicator will be changes in level of happiness. Since 2012, Sustainable Development Solutions Network has released the yearly [World Happiness Report](https://worldhappiness.report/archive/). This report uses the [Gallup World Poll (GWP)](https://www.gallup.com/178667/gallup-world-poll-work.aspx). This poll asks respondents, from nationally representative samples, to think of a ladder (Cantril ladder technique) with the best possible life for them being a ten and the worst life being a 0, and they are asked to rate their own life based on that scale. Each country's final rank is based entirely on the Gallup weights' survey scores to make the estimates representative. Even though the ladder score (mean of all rankings per country) is the primary indicator of happiness, the report also includes other descriptive statistics of that indicator, such as standard error of ladder score, upper whisker, and lower whisker. In addition to the ladder score, there are six factors (GDP per capita, life expectancy, generosity, social support, freedom, and corruption) used to estimate to what extent each contributes to making happiness evaluations in each country higher than in Dystopia. Dystopia is the hypothetical country with the lowest scores observed for each of the six factors. This imaginary country would have the world's least-happy people; therefore, no country can perform worse than Dystopia, establishing a benchmark against which all countries can be compared. 

In all, it is compelling to the democratic community to understand the current trends in democratic norms and civil liberties to act proactively in preventing the kinds of authoritarianism that threatens to destabilize the current world order and upend the global economic system. Analyzing trends over time may allow researchers to identify the signs and markers of democratic backslide at their nascent stages and recommend policy action to reverse the trend. History is often said to repeat itself, and understanding the causes of past events is crucial in understanding how future events may play out. 

Based upon the Global Happiness Report, we believe that the report index may be a useful indicator of periods of change in a country. We believe that it is plausible that the happier folks in a country are, the more likely they are to support the status quo, and the less likely it is to see a backsliding or improving democracy. Likewise, if the people of a nation are unhappy, it would be plausible to see evidence of changing democratic norms as unhappy people demand change. If this association is correct, then it may inform policy decisions to maximize the happiness of people (and improve the factors that contribute to the happiness score) in order to maintain stability in a country and around the world.

## **I. KMeans Clustering**

### **a. Re-Reading in the Data**

```python
# re-reading in all of the data for main democracy dataset 
# reading in the data
fname = "/content/drive/MyDrive/mlpp22/assignments/group1/V-Dem-CY-Core-v12.csv"
democracy = pd.read_csv(fname)

#getting only the columns we want 
cols = ['country_name', 'year', 'v2x_libdem', 'v2x_cspart', 'v2x_rule']
democracy = democracy[cols]

# reset the index
democracy = democracy.set_index('country_name')

# drop countries we don't need 
countries = ['Palestine/British Mandate', 'Baden', 'Bavaria', 'Modena', 'Parma', 'Saxony', 'Tuscany', 'Würtemberg', 'Two Sicilies', 'Hanover', 'Hesse-Kassel', 'Hesse-Darmstadt', 'Mecklenburg Schwerin', 'Papal States', 'Hamburg', 'Brunswick','Oldenburg', 'Saxe-Weimar-Eisenach', 'Nassau', 'Piedmont-Sardinia', 'Vanuatu', 'Seychelles', 'Sao Tome and Principe', 'Solomon Islands', 'Republic of Vietnam', 'Maldives', 'Palestine/West Bank', 'Eswatini', 'Palestine/Gaza', 'Somaliland', 'Zanzibar']
democracy = democracy.drop(countries)

# reset the index
democracy = democracy.reset_index()

# creating a dataframe for only LATAM countries
democracy = democracy.set_index('country_name')

# making a list of only Latin American countries 
DemocracyLatinAmerica = ['Mexico', 'Colombia', 'Brazil', 'El Salvador', 'Bolivia', 'Honduras', 'Peru', 'Argentina', 'Venezuela', 'Nicaragua', 'Chile', 'Costa Rica', 'Ecuador', 'Guatemala', 'Panama', 'Uruguay', 'Dominican Republic', 'Cuba', 'Guyana', 'Paraguay']

# Makign the Latin American country specific dataframe 
DemocracyLatinAmerica = democracy.loc[DemocracyLatinAmerica]

# resetting the index
DemocracyLatinAmerica = DemocracyLatinAmerica.reset_index()

# getting values only after 1960
DemocracyLatinAmerica1960 = DemocracyLatinAmerica['year'] >= 1960
DemocracyLatinAmerica1960 = DemocracyLatinAmerica[DemocracyLatinAmerica1960]
DemocracyLatinAmerica1960 = DemocracyLatinAmerica1960.reset_index()
DemocracyLatinAmerica1960 = DemocracyLatinAmerica1960.drop(columns='index')
DemocracyLatinAmerica1960.head(4)
```

### **b. Creating groupby dataframes over time for KMeans Cluster Analysis for our main features and target in the democracy dataframe**

```python
liberaldemocracyKMeans = DemocracyLatinAmerica1960.groupby(['country_name', 'year']).mean().unstack('year')['v2x_libdem'].dropna()
civilsocietystrengthKMeans = DemocracyLatinAmerica1960.groupby(['country_name', 'year']).mean().unstack('year')['v2x_cspart'].dropna()
ruleoflawstrengthKMeans = DemocracyLatinAmerica1960.groupby(['country_name', 'year']).mean().unstack('year')['v2x_rule'].dropna()

# extracting out the country names from one of the three main KMeans Cluster Datafraes that can serve for all KMeans anlayiss 
country = liberaldemocracyKMeans.index.values

## extracting out the years from the three main KMeans Cluster Datafrmaes that can serve for all Kemans analysis 
years = liberaldemocracyKMeans.columns.astype(int)

## getting the values to create numpy arrays for our data from our 3 main dataframes above for the KMeans clustering analysis

liberaldemocracyKMeansnp = liberaldemocracyKMeans.values
civilsocietystrengthKMeansnp = civilsocietystrengthKMeans.values
ruleoflawstrengthKMeansnp = ruleoflawstrengthKMeans.values

```

### **c. Standardizing the 3 main clustering dataframes created that contain our target and 2 main features respectively**

```python
# -- now let's cluster, but first (of course) standardize across columns to 
#    cluster based on trends in the time series for all 3 of our main variable dataframes 
liberaldemocracyKMeans_st = (liberaldemocracyKMeansnp - liberaldemocracyKMeansnp.mean(axis=1, keepdims=True)) / liberaldemocracyKMeansnp.std(axis=1, keepdims=True)
civilsocietystrengthKMeans_st = (civilsocietystrengthKMeansnp - civilsocietystrengthKMeansnp.mean(axis=1, keepdims=True)) / civilsocietystrengthKMeansnp.std(axis=1, keepdims=True)
ruleoflawstrengthKMeans_st = (ruleoflawstrengthKMeansnp - ruleoflawstrengthKMeansnp.mean(axis=1, keepdims=True)) / ruleoflawstrengthKMeansnp.std(axis=1, keepdims=True)
```

### **d. Using the "Elbow Method" to determine the optimal 'K'**




```python
# import the KMeans clustering algorithm 
from sklearn.cluster import KMeans
```

First -- I am employing the elbow method to find out the optimal K (# of clusters to use for our 3 variables of interest)  **From the plots below for our 3 variables of interest -- a K = 3 appears optimal for all 3 series (our target: liberal democracy; and 2 feauures (Rule of Law Strength & Civil Society Strength respectively)**

```python
# Liberal Democracy elbow method 
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(liberaldemocracyKMeans_st)
    distortions.append(kmeanModel.inertia_)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(K, distortions, 'bx-')
ax.set_xlabel('k')
ax.set_ylabel('Distortion')
ax.set_title('Elbow Method Liberal Democracy Optimal K')
plt.show()

# civil soceity strength elbow method 
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(civilsocietystrengthKMeans_st)
    distortions.append(kmeanModel.inertia_)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(K, distortions, 'bx-')
ax.set_xlabel('k')
ax.set_ylabel('Distortion')
ax.set_title('Elbow Method Civil Society Strength Optimal K')
plt.show()

# rule of law strength elbow method 
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(ruleoflawstrengthKMeans_st)
    distortions.append(kmeanModel.inertia_)

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(K, distortions, 'bx-')
ax.set_xlabel('k')
ax.set_ylabel('Distortion')
ax.set_title('Elbow Method Rule of Law Strength Optimal K')
plt.show()
```

**Figure 1: Presents the output plotted images of running the Elbow Method to find the optimal level of K clusters for our unsupervised algorithm. For all 3 series -- a K of 3 appears optimal as is evidenced by the breaking of the "elbow" along the x-axis 'k' values at the value of 3.**

### **e. Running our KMeans Clustering Algorithm**

---



---



```python

# -- cluster with 3 clusters

# instantiate & fit liberal democracy
kmliberaldemocracy = KMeans(n_clusters=3).fit(liberaldemocracyKMeans_st)

# instantiate & fit civil soceity strength 
kmcivilsocietystrength = KMeans(n_clusters=3).fit(civilsocietystrengthKMeans_st)

# instantiate & fit rule of law strength 
kmruleoflawstrength = KMeans(n_clusters=3).fit(ruleoflawstrengthKMeans_st)

```

### **d. Logistic Regression Machine Learning Output, Interpretation, & Visualization**

```python
#########################
#########################
### FITTING LOGIT FOR RULE OF LAW STRENGTH ON DEMOCRACY OUTCOME 
########################
########################

# -- let's fit a logistic regression model to this data
from sklearn.linear_model import LogisticRegression

# -- define features and target
XXrule = HappyDemocracy2015_2019['rule_of_law_strength'].values.copy()
YYDemocracy = HappyDemocracy2015_2019['Democracy_Outcome'].values.copy()

# -- initialize the model
lrrule = LogisticRegression()

# -- fit the model to data
lrrule.fit(XXrule.reshape(-1, 1), YYDemocracy)


# -- get predicted values
predrule  = lrrule.predict(XXrule.reshape(-1, 1))

# -- get model values
xvalrule = np.linspace(XXrule.min(), XXrule.max(), 100)
mmrule = lrrule.coef_[0][0]
bbrule = lrrule.intercept_[0]
zrule = xvalrule * mmrule + bbrule
modelrule = 1.0 / (1.0 + np.exp(-zrule))


#########################
#########################
### FITTING LOGIT FOR CIVIL SOCIETY STRENGTH ON DEMOCRACY OUTCOME 
########################
########################


# -- let's fit a logistic regression model to this data

# -- define features and target
XXcivil = HappyDemocracy2015_2019['civil_society_strength'].values.copy()
YYDemocracy = HappyDemocracy2015_2019['Democracy_Outcome'].values.copy()

# -- initialize the model
lrcivil = LogisticRegression()

# -- fit the model to data
lrcivil.fit(XXcivil.reshape(-1, 1), YYDemocracy)


# -- get predicted values
predcivil = lrcivil.predict(XXcivil.reshape(-1, 1))

# -- get model values
xvalcivil = np.linspace(XXcivil.min(), XXcivil.max(), 100)
mmcivil = lrcivil.coef_[0][0]
bbcivil = lrcivil.intercept_[0]
zcivil = xvalcivil * mmcivil + bbcivil
modelcivil = 1.0 / (1.0 + np.exp(-zcivil))


#########################
#########################
### FITTING LOGIT FOR CIVIL SOCIETY STRENGTH X RULE OF LAW STRENGTH INTERACTION TERM ON DEMOCRACY OUTCOME 
########################
########################

# -- let's fit a logistic regression model to this data

# -- define features and target
XXcivil_rule = HappyDemocracy2015_2019['Rule of Law X Civil Society'].values.copy()
YYDemocracy = HappyDemocracy2015_2019['Democracy_Outcome'].values.copy()

# -- initialize the model
lrcivil_rule = LogisticRegression()

# -- fit the model to data
lrcivil_rule.fit(XXcivil_rule.reshape(-1, 1), YYDemocracy)


# -- get predicted values
predcivil_rule = lrcivil_rule.predict(XXcivil_rule.reshape(-1, 1))

# -- get model values
xvalcivil_rule = np.linspace(XXcivil_rule.min(), XXcivil_rule.max(), 100)
mmcivil_rule = lrcivil_rule.coef_[0][0]
bbcivil_rule = lrcivil_rule.intercept_[0]
zcivil_rule = xvalcivil_rule * mmcivil_rule + bbcivil_rule
modelcivil_rule = 1.0 / (1.0 + np.exp(-zcivil_rule))


#########################
#########################
### FITTING LOGIT FOR HAPPINESS SCORE ON DEMOCRACY OUTCOME 
########################
########################


# -- let's fit a logistic regression model to this data

# -- define features and target
XXhappiness = HappyDemocracy2015_2019['Happiness Score'].values.copy()
YYDemocracy = HappyDemocracy2015_2019['Democracy_Outcome'].values.copy()

# -- initialize the model
lrhappiness = LogisticRegression()

# -- fit the model to data
lrhappiness.fit(XXhappiness.reshape(-1, 1), YYDemocracy)


# -- get predicted values
predhappiness = lrhappiness.predict(XXhappiness.reshape(-1, 1))

# -- get model values
xvalhappiness = np.linspace(XXhappiness.min(), XXhappiness.max(), 100)
mmhappiness = lrhappiness.coef_[0][0]
bbhappiness = lrhappiness.intercept_[0]
zhappiness = xvalhappiness * mmhappiness + bbhappiness
modelhappiness = 1.0 / (1.0 + np.exp(-zhappiness))





###################################
# PLOTTING THE LOGISTIC REGRESSIONS 
##################################

# -- plot the models and predictions
fig, ax = plt.subplots(1, 4, figsize=[30, 8])

# logistic regression predictions for rule of law strength
ax[0].scatter(XXrule, YYDemocracy, color="k", alpha=0.2)
ax[0].scatter(XXrule, predrule, s=15)
ax[0].plot(xvalrule, modelrule, color="darkorange", lw=3)
ax[0].set_xlabel("Rule of Law Strength", fontsize=15, weight='bold')
ax[0].set_ylabel("Outcome [1=Democracy, 0=Non-Democracy]", fontsize=15, weight='bold')
ax[0].tick_params(axis='x', labelsize=16)
ax[0].tick_params(axis='y', labelsize=16)

# logistic regression predictions for civil society strenth 
ax[1].scatter(XXcivil, YYDemocracy, color="k", alpha=0.2)
ax[1].scatter(XXcivil, predcivil, s=15)
ax[1].plot(xvalcivil, modelcivil, color="darkorange", lw=3)
ax[1].set_xlabel("Civil Society Strength", fontsize=15, weight='bold')
ax[1].set_ylabel("Outcome [1=Democracy, 0=Non-Democracy]", fontsize=15, weight='bold')
ax[1].tick_params(axis='x', labelsize=16)
ax[1].tick_params(axis='y', labelsize=16)

# logistic regression predictions for rule of law strength X Civil Society Strength 
ax[2].scatter(XXcivil_rule, YYDemocracy, color="k", alpha=0.2)
ax[2].scatter(XXcivil_rule, predcivil_rule, s=15)
ax[2].plot(xvalcivil_rule, modelcivil_rule, color="darkorange", lw=3)
ax[2].set_xlabel("Civil Society Strength X Rule of Law", fontsize=15, weight='bold')
ax[2].set_ylabel("Outcome [1=Democracy, 0=Non-Democracy]", fontsize=15, weight='bold')
ax[2].tick_params(axis='x', labelsize=16)
ax[2].tick_params(axis='y', labelsize=16)


# logistic regression predictions for plotting happiness
ax[3].scatter(XXhappiness, YYDemocracy, color="k", alpha=0.2)
ax[3].scatter(XXhappiness, predhappiness, s=15)
ax[3].plot(xvalhappiness, modelhappiness, color="darkorange", lw=3)
ax[3].set_xlabel("Happiness Score", fontsize=15, weight='bold')
ax[3].set_ylabel("Outcome [1=Democracy, 0=Non-Democracy]", fontsize=15, weight='bold')
ax[3].tick_params(axis='x', labelsize=16)
ax[3].tick_params(axis='y', labelsize=16)

fig.show()

ax[0].set_title('Panel 1: Rule of Law Strength In Sample Preds', weight='bold', fontsize=15)
ax[1].set_title('Panel 2: Civil Society Strength In Sample Preds', weight='bold', fontsize=15)
ax[2].set_title('Panel 3: Law X Civ Soc Strength In Sample Preds', weight='bold', fontsize=15)
ax[3].set_title('Panel 4: Happiness In Sample Preds', weight='bold', fontsize=15)
```

### **b. Extract out our Target & Feature Arrays** 

```python
# One hot encode year to add it into the model to see if there is temporal variation in the model have to make it string type
HappyDemocracy2015_2019['year'] = HappyDemocracy2015_2019['year'].astype(str)


# -- pull out the object IDs
country = HappyDemocracy2015_2019["country"]

# -- isolate the features
feat = HappyDemocracy2015_2019.drop(columns=['country', 'Democracy_Outcome', 'liberal_democracy'])

# -- one hot encode
feat = pd.get_dummies(feat)

# -- pull out the target variable
targ = HappyDemocracy2015_2019["liberal_democracy"]

feat
```

### **c. Running a base model -- Train & Test a Decision Tree Regression Model | Visualize | Interpretation**

```python
# -- create a random training and testing set
feat_tr, feat_te, targ_tr, targ_te = train_test_split(feat, targ, test_size=0.2, random_state=63)

# -- fit a decision tree classifier on the TRAINING data
dtc = DecisionTreeRegressor(random_state=63)
dtc.fit(feat_tr, targ_tr)

# -- assess the accuracy on the TRAINING data
pred_tr = dtc.predict(feat_tr)
mae_tr = mean_absolute_error(targ_tr, pred_tr)
print("On Average: Our predictinos for our training data were roughly = {0:.5f} units off".format(mae_tr))

# -- assess the accuracy on the TESTING data
pred_te = dtc.predict(feat_te)
mae_te = mean_absolute_error(targ_te, pred_te)
print("On Average: Our predictions for our testing data were roughly {0:.5f} units off".format(mae_te))
```

In our base model, we see that our training data fit the model the perfectly and our key metric of interest "mean absolute error" showed fit hte data perfectly in its predictions on the training data as we were 0.000 units off of the true value of what the actual values were. For our testing data, we were on average, about 0.0207 units off of accurately predicting our target variable. However, this model does seem suspect of overfitting given the spread in MAE between our training and testing data. ### **d. Decision Tree Regression Model Robustness -- Finding the decision tree regression model w/ the least mean absolute error**

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, feat_tr, feat_te, targ_tr, targ_te):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(feat_tr, targ_tr)
    preds_val = model.predict(feat_te)
    mae = mean_absolute_error(targ_te, preds_val)
    return(mae)
```

```python
## USING A FOR LOOP TO COMPARE MODELS WITH DIFFERENT MAX LEAF NODES 
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [2, 3, 4, 5, 6, 7, 8, 9, 10, 15, 25, 40, 50, 68, 69, 100, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, feat_tr, feat_te, targ_tr, targ_te)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %.5f" %(max_leaf_nodes, my_mae))
```

* From the function above --> the optimal decision tree regression model will have a **max_leaf_nodes of 9** given that it produced the lowest mean absolute error.
