**Travel and tourism is vital to the US economy -- especially because demand for travel stimulates economic activity in so many different sectors such as the hotel and restaurant industries in which millions of people depend upon for their livelihoods. Understanding the impact of the pandemic on travel is of vital importance for the US economy to return back to its normal pre-pandemic state. Have travel levels in 2021 returned to their pre-pandemic 2019 levels? Moreover -- what were the impacts of the pandemic on travel in general and on the economy overall? Answering these questions is of vital importance to understand the implications of the pandemic on the travel and tourism industries along with the overall health and direction of the US economy in general.**

**Brief Literature Review**

The impact of the covid 19 pandemic on the US commercial travel industry as it regards the nations airlines has took a tremendous toll. Unfortunately, due to travel restrictions imposed by governments along with general anxiety on behalf of travelers -- the airline and air travel industries were hit particularly [hard](https://www.bu.edu/cgs/2021/09/23/undergraduate-research-covid-19s-impact-on-the-airline-industry/). However, commercial flights were hit much more harder than non-commercial flights. This suggests to us that non-commercial flights generally kept up for all of the obscure reasons they usually do but commercial flights where passengers as rational economic actors make a decision to fly or not -- dropped significantly. What is even more interesting is the fact [that US domestic flights and travel were impacted much more significantly than international travel](https://www.bu.edu/cgs/2021/09/23/undergraduate-research-covid-19s-impact-on-the-airline-industry/).

A report from [Mckinsey](https://www.mckinsey.com/industries/travel-logistics-and-infrastructure/our-insights/back-to-the-future-airline-sector-poised-for-change-post-covid-19) showed that there was a roughly 40% drop in airline industry revenues in 2020 compared to 2019 data while projecting that air traffic will not return to 2019 levels until roughly 2024. Aside from the revenue drop -- the pandemic has significantly altered consumer behavior in ways which will stick with society where people and businesses realized they didn't even really need to do all of this traveling in the first place when business and or meetings could get done by simply hopping on a zoom call. Unfortunately for consumers as well -- the high levels of debt taken on behalf of airlines will also result in higher prices for consumers as companies need to be able to pay off their debt on schedule in order to stay afloat.

A report from the [OECD](https://www.oecd.org/coronavirus/policy-responses/covid-19-and-the-aviation-industry-impact-and-policy-responses-26d521c1/) also notes just how much the air travel industry does impact economies overall. The report notes that although the air travel industry represents a small share of total GDP -- it is closely linked to other economic sectors and the air travel industry is considered to be a key enabler of many other economic activities. Thus, it is crucial for us to explore and find out the impacts of the pandemic not only the travel industry itself -- but also for the overall US economy -- which is the aim of this analysis. 

The report from [Mckinsey](https://www.mckinsey.com/industries/travel-logistics-and-infrastructure/our-insights/back-to-the-future-airline-sector-poised-for-change-post-covid-19) does atleast offer some hope though in the fact that it states that we will as a country eventually recover. For instance, the report notes that leisure travel (pent up demand) and folks ready to get out will help to fuel the recovery aiding to dent the impacts of the altered consumer behavior that was formed as a result of the pandemic. At the end of the day, the same report from the [OECD](https://www.oecd.org/coronavirus/policy-responses/covid-19-and-the-aviation-industry-impact-and-policy-responses-26d521c1/) concludes that times will remain tough for the airline industry and passengers alike. Government intervention (strategically) will be key to helping both sides of the equation get back on the right footing.

**Reading in the main travel data and wrangling it to get it to form**

```python
 # read in the data 

url = "https://www.tsa.gov/coronavirus/passenger-throughput"
html = requests.get(url).content
travel = pd.read_html(html)
df = travel[-1]

#rename the columns 
df.columns = ['Date', '2021 Total', '2020 Total', '2019 Total']



# creating dates and total travel indepedent series for 2021 
dates2021 = df['Date']
dates2021

Total2021 = df['2021 Total']
Total2021

# creating reversed date ranges to line up with 2021 dates in original dataframe for both 2020 and 2019 data 
dTmp = pd.date_range(end='11/14/2020', periods=365, freq='D', closed = None)
dates2020 = [d for d in reversed(dTmp)]

dTmp2 = pd.date_range(end='11/15/2019', periods=365, freq='D', closed = None)
dates2019 = [d for d in reversed(dTmp2)]

# Extracing the 2020 and 2019 totals fro mthe original dataframe to put them into the new dataframe below
Total2020 = df['2020 Total']
Total2019 = df['2019 Total']

## creating a new dataframe that has all 2019 2020 and 2021 dates in it along with their corresponiding total passenger numbers 
DatesData192021 = pd.DataFrame(dates2020)

DatesData192021['2020 Total'] = Total2020
DatesData192021.columns = ['2020 Dates', '2020 Total Travel']

DatesData192021['2019 Dates'] = dates2019
DatesData192021['2019 Total Travel'] = Total2019

DatesData192021['2021 Dates'] = dates2021
DatesData192021['2021 Total Travel'] =Total2021



```

**Creating Mean Rolling Averages in the main airport travel dataframe**

```python
#ReversedDatesData192021 creating a new dataframe from the one above but this time with all values reversed from what was the 
# bottom is now the top in order to have the dates in ascending order and their corresponding values 

DatesData192021_r = DatesData192021[::-1]
DatesData192021_r

## Add in 7 and 30 day rolling averages into the dataframe for each year of data 2019 2020 and 2021 respectively 

DatesData192021_r['2019 RA7'] = DatesData192021_r['2019 Total Travel'].rolling(window=7).mean()
DatesData192021_r['2019 RA30'] = DatesData192021_r['2019 Total Travel'].rolling(window=30).mean()

DatesData192021_r['2020 RA7'] = DatesData192021_r['2020 Total Travel'].rolling(window=7).mean()
DatesData192021_r['2020 RA30'] = DatesData192021_r['2020 Total Travel'].rolling(window=30).mean()

DatesData192021_r['2021 RA7'] = DatesData192021_r['2021 Total Travel'].rolling(window=7).mean()
DatesData192021_r['2021 RA30'] = DatesData192021_r['2021 Total Travel'].rolling(window=30).mean()

# make 2021 dates a datetime 
DatesData192021_r['2021 Dates'] = pd.to_datetime(DatesData192021_r['2021 Dates'])


```

# **Statistical Analysis/Predictions**

**Total Travel Prediction/Statistical Analysis**

Creating the dataframe for the analysis 

```python
# creating the dataframe for the analysis 

TravelNum = ts_travel['Total Travel']
TravelDate = ts_travel['Date']


predictionTravel = pd.DataFrame(TravelNum)
predictionTravelDate = pd.DataFrame(TravelDate)

TravelPredictions = predictionTravel.merge(predictionTravelDate, left_index=True, right_index=True)

TravelPredictions = TravelPredictions.reset_index()
TimeCounter1 = TravelPredictions['index']
TravelPredictions.columns = ['TimeCounter1', 'TravelNum', 'TravelDate']
TravelPredictions = TravelPredictions.sort_values('TravelDate', ascending=True)
TravelPredictions = TravelPredictions.dropna()
TravelPredictions.head(2)
```

Building the linear model

```python
# -- first build the model
model = sm.ols("TravelNum ~ TimeCounter1", data=TravelPredictions)

# -- now fit the model to the data
result = model.fit()
```

Model results 

```python
# -- summarize the model fit
print('Table 1')
print(result.summary())
```

From the results above in the OLS output as shown in Table 1: we see that our predictor variable in attempting to explain the variation in 'Total Travel numbers', that our 'TimeCounter1' predictor was highly statistically significant at the p < .01 level. However, the results remain biased as our data is not normally distributed according to the Jarque-Bera test in which we reject the null hypothesis of normality in the in the data. 

Model predictions for the linear model 

```python
# -- calculate the model "prediction" for year (our x values)
pred = result.predict(TravelPredictions) 
```

Building the quadratic model and corresponding results 

```python
# -- build the quadratic model
model2 = sm.ols("TravelNum ~ TimeCounter1 + np.power(TimeCounter1, 2)", data=TravelPredictions)


# -- now fit the model to the data
result2 = model2.fit()

# -- summarize the fit
print('Table 2')
print(result2.summary())
```

Table 2 above presents the results of our quadratic fit for our linear model in attempting to estimate the variation in total air travel numbers by our key predictor (time). We see that according to Table 2, our predictor for the quadratic effects has a coefficient value of 4.80 and is highly statistically significant suggesting to us that as time goes by -- we should see a corresponding increase in total air travel numbers as well. However, once more, although table 2 that fits the quadratic model as well as the linear model relative to table 1 that only plots the linear model -- our results still remain biased as our data is not normally distributed as we reject the null hypothesis of normality in our data as shown by our Jarque Bera statistic of 32.486.

**Plotting the Results for travel number predictions** 

```python
# -- predict the GDP per capita with the new model
pred2 = result2.predict(TravelPredictions)
```

# **Problem Solution/Discussion of Results**

The aim of this analysis was to explore the impact of the COVID-19 impact on air travel traffic in the United States and whether or not air travel numbers have returned to their pre-pandemic 2019 levels alongside analyzing the economic implications of the pandemic with respect to the correlations (if any) between travel numbers picking back up and the overall US economic recovery. 

**Data Analysis**

For one, we have conclusively demonsrated through data analysis and visualization techniques that air travel numbers have recovered from their 2020 lows in 2021 and are continuing to improve compared to when the pandemic initially struck in March of 2020. However, air travel numbers still remain depressed compared to their 2019 counterparts. 

We also demonsrated that there does appear to be some sort of weekly periodicity occuring in the data that is more visible to see in the 2019 and 2021 series and especially relative to the 2020 series where the impact of the pandemic is heavily imprinted on the series. We showed with autocorrelations with AR processess 1:7 that indeed a lag of 7 did have the strongest correlation of each 2019 and 2021 series with itself relative to lags 1:6 suggesting to us that travel picks up at the beginning of the week but dies down during the week when folks most likely arrive to their destination but picks back up again in about a week's time frame due to either folks returning to their destinations or for folks beginning travel anew in a new week. In order to examine the weekly periodicity a little bit more in depth with actual days of the week -- we created a heat map and stacked bar plot of the day of the week and its total travel numbers and the respective year. We showed that in both plots -- Sunday appears to be the day with the busiest and most extensive amount of travel which suggests to us that folks are often leaving and or returning to their destinations/origins on Sunday which makes intuitive sense given that Monday starts the normal work week for most folks. 

We also demonstrated that air travel numbers and the S&P 500 index (which serves as a correlate for the market) dropped in tandem with eachother when the pandemic hit in March of 2020 and subsequently began recovering together as well. However, the recovery of the market and economy has been far more robust than that of travel numbers suggesting to us the market may be a bit detached from what is actually happening on main street. We also conducted correlation analysis and demonsrtated that there was hardly any correlation between air travel numbers and the broader market (which serves as a correlate for the economy) albeit the series due appear to move in tandem with eachother with respect to the steep drop and subsequent recovery in activity.  

**Statistical Analysis/Forecasting**

With regards to statistical analysis and forecasting -- we showed that the totai air travel series 2019:2021 data is best fit via a quadratic model and that air travel is expected to recover to new highs stemming from our estimations but the results of this analysis are trivial as the series violates several gauss-markov assumptions and is far from a normal distribution.

We also demonstrated that in terms of the economic recovery -- the market as measured via the S&P 500 index predicted out into Nov 2024 is expected to reach new all time highs at an index price at around 6,250. Thus, regardless of what happens with air travel demand numbers if they ever get back to their 2019 levels or not -- the market will keep reaching new all time highs. 

**Geospatial component of the analysis**

Finally, with respect to the interactive visualization component of our analysis, we have shown that after analyzing 15 different prominent US airports across the United States and the % change in airport travel traffic from 2019 to 2020 -- that airports that were located in conservative states such as Texas and Florida and their corresponding airports -- saw the lowest % change difference between airport travel traffic from 2019 to 2020 when the pandemic hit. On the other hand, we demonstrated that those airports that were in highly liberal states such as California and New York -- that the corresponding airports in these liberal states experienced some of the greatest % change differences between the total airport travel traffic from 2019 to 2020. Overall, this finding appears to mirror political trends and dynamics where we know conservative states implemented the least restrictive COVID policies and liberal states implemented the most restrictive COVID policeis -- thus and correspondingly, airports in conservative states saw the least amount of percentage change in their total traffic numbers from 2019 to 2020 and airports in liberal states saw the most amount percentage change in their total traffic numbers from 2019 to 2020.
