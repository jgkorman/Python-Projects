```python
# load in necessary packages for anlaysis 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
!pip install --upgrade xlrd
```

```python
# pulling the datasource from the web
fname = "https://api.worldbank.org/v2/en/indicator/FP.CPI.TOTL.ZG?downloadformat=excel"
# reading in the data source
infl = pd.read_excel(fname, skiprows=3)
```

##**2. Data Wrangling**

```python
# make all columns into a list 
cols = infl.columns.to_list()
# copy and paste the columns list format into a new variable called "cols" --> and rearange the country name to the 4th column to index that value 
cols = ['Country Code',
 'Indicator Name',
 'Indicator Code',
 'Country Name',
 '1960',
 '1961',
 '1962',
 '1963',
 '1964',
 '1965',
 '1966',
 '1967',
 '1968',
 '1969',
 '1970',
 '1971',
 '1972',
 '1973',
 '1974',
 '1975',
 '1976',
 '1977',
 '1978',
 '1979',
 '1980',
 '1981',
 '1982',
 '1983',
 '1984',
 '1985',
 '1986',
 '1987',
 '1988',
 '1989',
 '1990',
 '1991',
 '1992',
 '1993',
 '1994',
 '1995',
 '1996',
 '1997',
 '1998',
 '1999',
 '2000',
 '2001',
 '2002',
 '2003',
 '2004',
 '2005',
 '2006',
 '2007',
 '2008',
 '2009',
 '2010',
 '2011',
 '2012',
 '2013',
 '2014',
 '2015',
 '2016',
 '2017',
 '2018',
 '2019',
 '2020',
 '2021']
```

```python
# boolean indexing to get all rows and only the 4th column onward 
infl = infl[cols].iloc[:, 3:]
infl
```

```python
# setting the index to country name --> and then transposing it to make countries all the columns and years going down on all rows -- converting data from wide to long format
infl = infl.set_index('Country Name').T
infl
```

```python
# resetting the index
infl = infl.reset_index()
infl
```

```python
# rename the column that contains our years to "Year" from its prior name "index"
infl = infl.rename(columns={'index': 'Year'})
infl
```

```python
# drop the index name -- our data is now formatted perfectly 
infl = infl.rename_axis(None, axis=1)
infl
```

```python
# taking only a subset of countries and columns to analyze inflation data on 
infl = infl[['Year', 'United States', 'United Kingdom', 'Spain', 'Colombia', 'Chile']]
infl
```

```python
# checking the datatypes of our variables 
infl.dtypes
```

```python
# changing the variable Year to an integer type 
infl['Year'] = infl['Year'].astype(int)
# re-checking our data types 
infl.dtypes
```

```python
# plotting inflation via scatter plot for a subset of our countries under study
# creatinga figure and an axis
fig, ax = plt.subplots(figsize=(10 ,5))

# plotting the data for three countries w/ different marker styles and colors 
infl.plot('Year', 'United States', marker='o', kind = 'scatter', label = 'United States', ax=ax)
infl.plot('Year', 'Chile', marker = 'x', kind = 'scatter', c='black', s = 100, label ='Chile', ax=ax)
infl.plot('Year', 'Colombia', kind='scatter', marker = 's', c='red', s=10, label='Colombia', ax=ax)

# setting the x and y labels and adjusting their fontsize
ax.set_xlabel('Year', fontsize=14, weight='bold')
ax.set_ylabel('Inflation [%]', fontsize=14, weight='bold')

# adding in a title 
ax.set_title('Inflation [%] 1960-2021', weight='bold', fontsize=20)

# annotatin the graph with text and an arrow to demonstrate the outlier clearly on the graph 
ax.annotate('Outlier', xy = (1976, 504),
            xytext = (1996, 502),
            arrowprops = dict(facecolor = 'blue'))




```
