Ride-hailing apps have taken over the traditional and once robust use of taxis in urban environment. The impact has been tremendous to say the least. First, before we dive into their impact on taxis in traditionally urban environments -- it is important to note that [studies](https://www.hindawi.com/journals/jat/2020/8831674/) have found that ride hailing apps have actually decreased private car travel in some contexts. However, it has also increased private vehicle ownership in other contexts -- this is especially the case in car [dependent slow growth cities](https://www.cmu.edu/news/stories/archives/2021/january/impact-rideshare-apps.html). Thus, taxis in urban environments are being impacted by a double whammy. On the one hand, ride-hailing apps have taken away from the traditional demand of taxis and shifted demand to their own respective domains. On the other hand, ride hailing apps have even increased car owneership and have caused more drivers to be on the road in order to make an extra amount of money to support living expenses -- further hurting the demand of taxi ridership in urban environments. An [investopedia](https://www.investopedia.com/articles/investing/110614/taxi-industry-pros-cons-uber-and-other-ehail-apps.asp) article notes that at one point in Manhattan -- yellow taxis ruled the streets. Now, however, just as recently as 2020, there were nearly 4 times as many ride-hailing vehcicles on the streets relative to taxis. The article goes on to note that ride-sharing apps such as Uber and Lyft have severely disrupted the taxi industry in nearly all formats. This has had the impact of replacing once steady jobs with gig economy work. [This is evident by the fact that taxi drivers use to be able to build up a nest egg and eventually purchase their own cab vehicle too](https://www.forbes.com/sites/michaelgoldstein/2018/06/08/uber-lyft-taxi-drivers/?sh=21556d5059f0) -- these golden days are long gone with the impact that ride-hailing apps have had on the industry. In fact, as a consequence, we have seen an increase in suicide rates amongst taxi cab drivers as the once middle class life they could reach doing their job has now become out of reach. Moreover, taxis have traditionally been farily priced. With ride share apps, however, consumers are starting to realize they are paying a lot more for a simple ride. Moreover, as ride sharing apps have attempted to push the price of travel down (although not consistently) it has cut into taxi profits as well -- thereby creating almost a 'race to the bottom' phenomena effect. 

Ride sharing apps are also not only taking away what one would have thought 'busyness' in the streets through the shrinking of taxis and private vehicles alike in the city streets -- but they are actually clogging streets more now than ever before. [They are having a deleterious impact in terms of traffic control](https://usa.streetsblog.org/2021/01/23/study-taking-an-app-taxi-more-than-doubles-your-roadway-impact/). At the end of the day, the dislocation and disruption of the taxi industry on behalf of the ride-hailing industry will have effects that will reverberate across workers for years to come. Those impacts are being acutely felt especially in urban contexts -- as taxi drivers lose their livelihoods and the streets become more and more congested... it is a lose lose for evrerybody..

```
Taxisubridehail = taxisub.xs('Ride-hailing apps', level='parent_type', axis=1).reset_index()
Taxisubtaxi = taxisub.xs('Taxis', level='parent_type', axis=1).reset_index()
Mergedtaxisub = Taxisubridehail.merge(Taxisubtaxi, left_on='date', right_on='date')
Mergedtaxisub
```

```
taxisubRA = taxisub.rolling(365, center=True).mean()
```

```

MergedRA = Ridehailingparent.merge(Taxiparent, left_on='date', right_on='date')
```

**Autocorrelation -- Determining Periodicity**

```
# -- isolate short time scale behavior  ## -- This detrends the data it is a form of ridding the data of stationarity
taxi_detrended = taxisub - taxisubRA
taxi_detrended
```

```
fig, ax = plt.subplots(figsize=(15, 5))
taxi_detrended.plot(y='trips', legend=False, color=['green', 'blue'], ax=ax)
ax.grid(axis="x", which='major', lw=2) # to set the grid lines on either the major ticks or the minor tick marks using which method
ax.grid(axis="x", which='minor', lw=0.5) 
ax.legend()

ax.set_xlabel("")
ax.set_ylabel("Detrended Trips ")
fig.show()
```
